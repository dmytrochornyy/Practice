### **Завдання 1 (1,5 бали)**

У 1939 році друком вийшов роман американського письменика Ернеста Вінсента Райта "Ґетсбі". Особливу увагу публіки він здобув через те, що письменник жодного разу не використав у романі літеру "e" (роман складається з більше ніж 50,000 слів), яка є найбільш вживаною літерою в англійській мові.

Порахуйте кількість входжень кожної літери у текст роману. Виведіть список літер англійського алфавіту, а також інформацію про те, скільки відсотків від усіх літер тексту складає кожна літера алфавіту.

Зауваження:

1. Програма має ігнорувати будь-які символи, окрім літер.

2. Переведіть усі літери у верхній або у нижній регістр.

3. Під час виведення відсортуйте літери у порядку зменшення їхнього відсотка. Літера "e" має стояти на останньому місці у цьому списку.

4. Виведіть перші 5 та останні 5 літер відсортованого списку разом з відсотковими значеннями, що їм відповідають.

```python
PTH = 'files\gadsby.txt'
ALPHABET = set('abcdefghijklmnopqrstuvwxyz')
letters_dict = {i:0 for i in ALPHABET}
p_dict = {i:0 for i in ALPHABET}
with open(PTH) as f:
    txt = "".join(f.readlines()).lower()
for letter in txt:
    if letter in ALPHABET:
        letters_dict[letter] += 1
all_letters = sum(letters_dict.values())
for letter in p_dict.keys():
    p_dict[letter] = (letters_dict[letter] / all_letters) * 100
to_print = {k: v for k, v in sorted(p_dict.items(), key=lambda item: item[1], reverse=True)}
print("First 5:\n")
counter = 0
for k in to_print:
    if counter == 5:
        break
    print(f"{k}: {round(to_print[k], 2)}%")
    counter += 1
print()
print("Last 5:\n")
for k in to_print:
    if counter == 29:
        break
    elif counter >= 23:
        print(f"{k}: {round(to_print[k], 2)}%")
    counter += 1
print()
print("All:\n")
for k in to_print:
    print(f"{k}: {round(to_print[k], 2)}%")
```

### **Завдання 2 (1,5 бали)**

Створіть CSV-файл, що буде містити від 5 до 10 пісень (творів) вашого улюбленого музиканта. Назва файлу - ім'я музиканта (композитора) або назва гурту.

1. Обов'язково додайте до файлу рядок з назвами двох колонок (Song, Year) або (Composition, Year).

2. Після запису файлу роздрукуйте назву файлу та усі наявні у файлі рядки (включно із заголовками).

3. Для запису та зчитування рядків скористайтесь функціями csv.DictWriter() та csv.DictReader().

```python
PTH = "files\Eminem.csv"

import csv
songs = [
    {
        "Composition": "Lose yourself",
        "Year": "2002"
    },
    {
        "Composition": "’Till I Collapse",
        "Year": "2002"
    },
    {
        "Composition": "Stan",
        "Year": "2000"
    },
    {
        "Composition": "Rap God",
        "Year": "2013"
    },
    {
        "Composition": "Venom",
        "Year": "2018"
    }
]

with open(PTH, 'w', newline='') as file:
    columns = ['Composition', 'Year']
    writer = csv.DictWriter(file, fieldnames=columns)
    writer.writeheader()
    for i in songs:
        writer.writerow(i)

f = open(PTH, 'r')
with f:
    reader = csv.DictReader(f)
    print(*list(reader.fieldnames))
    for row in reader:
        print(dict(row))
```

3. Визначте кількість наявних категорій (кількість елементів, що відповідають ключу "categories").

4. Виведіть посилання на фотографію 000000000001.jpg ("coco_url"), висоту цієї фотографії ("height"), ширину ("width") та її ідентифікатор ("id").

5. Опис структури JSON-файлу для набору фотографій COCO можна знайти за посиланням: https://cocodataset.org/#format-data

6. Щоб перевірити правильність результатів, у VS Code встановіть розширення JSON та відкрийте файл image_info_test-dev2017.json. Розширення автоматично визначає кількість елементів у кожному полі.

```python
import json
PTH = r"files\annotations\image_info_test-dev2017.json"

with open(PTH) as file:
    dt = json.load(file)

n_of_photos = len(dt["images"])
print("Number of photos in dataset:", n_of_photos)
n_of_categories = len(dt["categories"])
print("Number of categories in dataset:", n_of_categories)

for i in dt['images']:
    if i['file_name'] == "000000000001.jpg":
        url = i["coco_url"]
        height = i["height"]
        width = i["width"]
        _id_ = i["id"]
print("URL of 000000000001.jpg:", url)
print("Height of 000000000001.jpg:", height)
print("Width of 000000000001.jpg:", width)
print("id of 000000000001.jpg:", _id_)
```
