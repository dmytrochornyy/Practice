Функція random_matrix() призначена для створення квадратної матриці випадкових чисел. Її змінювати непотрібно.

Приклад використання методу permutations() для створення списку перестановок також подається у комірці з кодом.

1) Створити 3 функції: функцію створення списку перестановок, функцію підрахунку добутків та функцію підрахунку загальної суми.

2) Кожна функція має супроводжуватись рядком документації, що буде коротко описувати її призначення.

3) Розмірність квадратної матриці має вводитись користувачем з клавіатури. Підрахунки виконувати лише в тому разі, якщо користувач введе ціле додатне число.

4) Загальний результат повинен підраховуватись за допомогою композиції функцій.

За бажання можна перевірити правильність роботи програми за допомогою функції np.linalg.det(a).

```
import numpy as np
import itertools

def random_matrix(dim):
    """
    The function generates dim x dim array of integers
    between 0 and 10.
    """
    matrix = np.random.randint(10, size = (dim, dim))
    return matrix


def list_of_permutations(matrix):
    """
    The function creates an array with all posible permutations of given matrix.
    """
    return list(itertools.permutations(range(len(matrix))))


def multiply_func(matrix, permutation):
    """
    The function creates an array with all posible permutations of given matrix.
    """
    result = 1
    for i in range(len(permutation)):
        result *= matrix[i][permutation[i]]
    return result


def inversions_sign(permutation):
    """
    The function counts if permutation is even or odd, and gives us a sign.
    """
    counter = 0
    for i in permutation:
        for j in permutation:
            if i < j and permutation.index(i) > permutation.index(j):
                counter +=1
    return (-1)**counter


def determinant(matrix):
    """
    The function counts sum of all 'members' of determinat.
    """
    result = 0
    for permutation in list_of_permutations(matrix):
        result += inversions_sign(permutation) * multiply_func(matrix, permutation)
    return result

def input_int():
    """
    The function gets an integer from keyboard.
    """
    inp = input("Enter matrix size as a positive integer: ")
    try:
        if int(inp) > 0:
            return int(inp)
    except ValueError:
        print("It`s not an positive integer. Please, enter positive integer.")
        return input_int()

def test():
    """
    Gets a random matrix of given size and prints this determinant.
    """
    matrix = random_matrix(input_int())
    print("Your matrix:", *matrix, sep="\n")
    print("Determinant of this matrix: ", determinant(matrix))


test()
```
